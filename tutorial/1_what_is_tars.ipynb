# 1. Tarsとはなにか

Tarsは生成モデル，特に深層生成モデルを簡単に実装するためのライブラリです．

確率モデリングのためのライブラリは様々ありますが，代表的なのは（Py）Stan，PyMC3，Edwardです．

これらのライブラリと比較すると，Tarsには深層生成モデルを実装するための工夫として次の違いがあります．

## 1) 深層ニューラルネットワークを確率分布によって隠蔽する．

例えばEdwardでは，確率分布だけではなく深層ニューラルネットワークも混ぜて記述することができます．これによって，VAEなどの深層生成モデルも実装することができます．

しかし， それぞれの確率分布からサンプリングすることはできませんし，それぞれの尤度を評価することもできません．

そこでTarsでは深層ニューラルネットワークを確率分布で隠蔽するように記述します（下図，詳しくは次回説明します）．

<img src="images/overview.png" width=50%>

これによって，外部からは確率分布の操作にしかみえなくなるので，簡単に尤度計算やサンプリングができるようになります．

<img src="images/sampling.png" width=30%>

この方法は，深層ニューラルネットワークをモジュールとして見立てて，確率的に組み合わせて実装する方法といえます．

## 2) 同時分布を記述して確率モデルを設計する．

多くの確率モデリングライブラリが，確率過程を記述する形で実装しています．

これはパラメータをMCMCなどで計算する際には効率的ですが，深層生成モデルでは変分下界を求める必要があり，各分布の尤度を計算しなければなりません（下図）．

<img src="images/lowerbound.png" width=50%>

その際に，確率過程の形よりも，同時分布で与えるほうが，複雑な確率モデルを容易に計算できます．

Tarsでは，確率モデリングライブラリとして初めて**各確率分布の積の形で同時分布を明示的に記述するAPI**を採用しました（こちらも詳しくは次回説明します）． 

これによって，上記に書いたように，同時分布の形を明示的に定義することができますが，もう一つの利点として，確率分布の積を新たに1つの確率分布として扱うことができるということがあります．

例えば，確率分布として$p(x|z)$と$p(z|y)$を定義し，その後$p(x,z|y)=p(x|z)p(z|y)$として新たに確率モデルをつくると，$p(x,z|y)$を（どのような確率分布で構成されているかを無視して）一つの確率分布として尤度計算やサンプリングができるようになります．

<img src="images/sampling2.png" width=30%>

これによって，アルゴリズム側では，分布の形を一切気にせずに計算することができます．

## Tarsではできないこと

多くの確率モデリングライブラリとは異なり，Tarsは現時点ではベイズ推論はできません．
これは，本ライブラリの目的が深層生成モデルの実装にあるからです．

また， 計算方法についても確率的変分推論や最尤推定などしかサポートしておらず，サンプリングによる計算方法は実装されていません．

これらについては，今後サポートしていきたいと思います．
